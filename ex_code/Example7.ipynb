```python
import scipy as sp
import math as mt
import numpy as np
from scipy.integrate import quad
import math as mt
import matplotlib.pyplot as plt
import matplotlib as mplt

print(sp.__version__)
print(np.__version__)
print(mplt.__version__)
```

```python
r = 0.10
sig = 0.2
S0 = 100
T= 10

M=10000

dt=T/M

S = np.zeros(M+1)

xi = np.zeros(M)

xi = np.random.normal(size = M)
#print(np.min(xi))


S[0] = S0

for k in range(0 , M):
    S[k+1] = S[k] + S[k] * r * dt + S[k] * sig * mt.sqrt(dt) * xi[k]





```

### Plot

```python
plt.plot(range(M+1), S, color='blue')

plt.xlabel('\nNumber of steps')
plt.ylabel('S t')
plt.title('\n')
#you need to change x-axis limit depending on your result
plt.xlim((0,M+10))
#you need to change y-axis limit depending on your result
plt.ylim((np.min(S),np.max(S)))

plt.show()
```
